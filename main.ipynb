```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# Define paths
data_2 = 'Data/US_2/t3US51_738969_3.jpg'
label_2 = 'Data/Labels_2/t3Label36_738969_3.png'

data_1 = 'Data/US_Test_2023April7/t3US12_738983_1.jpg'
label_1 = 'Data/Labels_Test_2023April7/t3Label23_738983_1.png'


# Load images using OpenCV (BGR format)
img1 = cv2.imread(label_2)
img2 = cv2.imread(label_1)
img3 = cv2.imread(data_1)
img4 = cv2.imread(data_2)

# Convert from BGR to RGB (for proper color display in Matplotlib)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)

# Create a figure with 2 rows and 2 columns
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Plot images
axes[0, 0].imshow(img1)
axes[0, 0].set_title("Label 2")
axes[0, 0].axis("off")  # Hide axis

axes[0, 1].imshow(img2)
axes[0, 1].set_title("Label 1")
axes[0, 1].axis("off")  # Hide axis

axes[1, 0].imshow(img3)
axes[1, 0].set_title("Data 1")
axes[1, 0].axis("off")  # Hide axis

axes[1, 1].imshow(img4)
axes[1, 1].set_title("Data 2")
axes[1, 1].axis("off")  # Hide axis

# Adjust layout and show the images
plt.tight_layout()
plt.show()
```
